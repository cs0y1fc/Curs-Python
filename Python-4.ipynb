El Objeto iloc (Integer Location) permite acceder a filas/columnas de un dataframe utilizando los indices numericos.

El indice de la primera fila es 0

El indice de la primera columna es 0

iloc[fila,columna]

```python
# Lectura

trimestre.iloc[0,0]

trimestre.iloc[0,1]

# Escritura
trimestre.iloc[0,1] = "Quesito frito"
trimestre
```

```python
# Con iloc puedo ver un conjunto de registros y ademas por el numero de columnas que quiera y en el orden que quiera

# 2 puntos: Cuando se pide solo 1 rango, se ponen : en el otro

trimestre.iloc[20:30,]
# Ver solo la columna 1, para todas las filas
trimestre.iloc[:,[1]]
# Ver solo las columnas 3,6,1 y 2, para todas las filas (en este orden)
trimestre.iloc[:,[3,6,1,2]]
```
