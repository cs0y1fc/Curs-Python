el dataframe trimestre. Realizar las transformaciones necesarias para resolver
los siguientes enunciados:
- Crear la columna Amount = Unit Price * Quantity
- Crear la columna Enterprise = "Delta Enterprise"
- Crear la columna Discount que sera o bien el 3,35% sobre Amount si el pais
  es FRANCE o GERMANY o bien 0
- Ver en un texto la suma del Discount es X
- Ver el promedio de Amount de los pedidos cuya Category = Beverages

```python
#Primero instalar openpyxl, esto es a nivel de PC
!pip install openpyxl
#Importar la libreria pandas, es a nivel kernel
import pandas as pd
#Cuando abrimos un fichero de python, su kernel esta vacio
#no tiene importadas, ni las librerias, ni variables, ni funciones, etc...
```

```python
#Cargar el Agosto
agosto2014 = pd.read_excel(r"C:\Users\formacio\Downloads\VentasGroucery_Meses.xlsx", 
                           sheet_name="Ventas Agosto 2014")
#agosto2014
septiembre2014 = pd.read_excel(r"C:\Users\formacio\Downloads\VentasGroucery_Meses.xlsx",
                           sheet_name="Ventas Septiembre 2014")
#septiembre2014
octubre2014 = pd.read_excel(r"C:\Users\formacio\Downloads\VentasGroucery_Meses.xlsx", 
                           sheet_name="Ventas Octubre 2014")
#octubre2014
#Juntar los 3 meses:
trimestre = pd.concat([agosto2014,septiembre2014,octubre2014],ignore_index=True)
trimestre.head(5)
trimestre.tail(5)
#Calcular la columna nueva Amount
trimestre['Amount'] = trimestre['Unit Price']*trimestre['Quantity']
trimestre
#Columna enterprise
trimestre['Enterprise'] = "Delta Enterprise"
trimestre
#Calcular el Descuento, lo voy a hacer con una función
def descuento(Country):
    if (Country=="France") or (Country=="Germany"):
        return 0.0335
    else:
        return 0
#Aplicar el descuento a la columna
trimestre['Discount'] = round(trimestre['Amount']*trimestre['Ship Country'].apply(descuento),2)
trimestre
#Ver la suma total de Discount
print("El importe total de descuento es",round(trimestre['Discount'].sum(),2))
#Ver el promedio de Amount de Beverages
FCB = trimestre['Category']=="Beverages"
resultado = round(trimestre[FCB].Amount.mean(),2)
resultado = round(trimestre[FCB]['Amount'].mean(),2)
print("El promedio de Amount de la categoria Beverages es", resultado,"€")
#Suma de Amount por Seafood y con la Quantity entre 25-75
FCSQ2575 = (trimestre['Category']=="Seafood")&(trimestre['Quantity']>=25)&(trimestre['Quantity']<=75)
resultado = trimestre[FCSQ2575]['Order ID'].count()
print("La cantidad de pedidos de Seafood, Quantity 25-75 es", resultado)
```

la suma total de Amount de los 5 meses.

```python
noviembre2014 = pd.read_excel(r"C:\Users\formacio\Downloads\VentasGroucery Nov-Dic.xlsx",
                           sheet_name="Ventas Noviembre 2014")
diciembre2014 = pd.read_excel(r"C:\Users\formacio\Downloads\VentasGroucery Nov-Dic.xlsx", 
                           sheet_name="Ventas Diciembre 2014")
novdic2014 = pd.concat([noviembre2014, diciembre2014])
#En este dataframe de 2 meses voy a calcular Amount
novdic2014['Importe'] = novdic2014['Precio Unidad'] * novdic2014['Cantidad']
novdic2014

if len(trimestre.columns)!=len(novdic2014.columns):
    print("El numero de columnas es diferente - trimestre",len(trimestre.columns),
          'novdic2014 tiene',len(novdic2014.columns))
    #ajustar el dataframe trimestre para que tenga solo las 10 columnas
    trimestre = trimestre.iloc[:,0:10]
    print(len(trimestre.columns))
#Cambiar el nombre de las columnas, para que sean las mismas
novdic2014.columns = trimestre.columns
novdic2014.columns
cincomeses = pd.concat([trimestre,novdic2014],ignore_index=True)
#Calcular el Amount
resultado = cincomeses['Amount'].sum()
print("La suma de Amount de los 5 meses es", resultado)
```

```python
#Como agrupar por una dimensión, dimensión = es todo aquello por lo que 
#queremos medir, ejemplo Ship_Country, Category, Ship_City.
#la metrica es la medición de la información, suma, el promedio, mas grande
#mas pequeño, la cantidad
#dimensión Ship_Country y la metrica la suma de Amount
#Germany 125788
#France   88656
#Italy    75222
cincomeses.groupby(['Ship Country']).Amount.sum()

cincomeses.groupby(['Ship Country']).Amount.count()

#Y si fuera por mas de una dimensión
cincomeses.groupby(['Ship Country','Category']).Amount.count()
```
